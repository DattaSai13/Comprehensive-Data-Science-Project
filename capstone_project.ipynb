```python
from src.preprocessing import load_datasets, preprocess_churn, preprocess_house, preprocess_sales
from src.train_model import train_churn_model, train_house_model, train_sales_model
from src.evaluate import evaluate_churn, evaluate_regression

# Load datasets
sales, house, churn = load_datasets()

# Preprocess
X_train_c, X_test_c, y_train_c, y_test_c = preprocess_churn(churn)
X_train_h, X_test_h, y_train_h, y_test_h = preprocess_house(house)
X_train_s, X_test_s, y_train_s, y_test_s = preprocess_sales(sales)

# Train models
churn_model = train_churn_model(X_train_c, y_train_c)
house_model = train_house_model(X_train_h, y_train_h)
sales_model = train_sales_model(X_train_s, y_train_s)

# Evaluate
evaluate_churn(churn_model, X_test_c, y_test_c)
evaluate_regression(house_model, X_test_h, y_test_h, "House Model")
evaluate_regression(sales_model, X_test_s, y_test_s, "Sales Model")
```
